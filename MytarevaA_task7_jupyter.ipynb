```python
# КОНСОЛЬНЫЙ КАЛЬКУЛЯТОР (гарантированно работает везде)

from datetime import datetime
import os

class ConsoleCalculator:
    def __init__(self):
        self.history_file = "calculations_history.txt"
        print(f"Файл истории: {os.path.abspath(self.history_file)}")
        self.run()
    
    def save_history(self, expr, result):
        """Сохраняет операцию в файл"""
        with open(self.history_file, 'a', encoding='utf-8') as f:
            timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            f.write(f"[{timestamp}] {expr} = {result}\n")
    
    def show_history(self):
        """Показывает историю"""
        print("\n" + "="*50)
        print("ИСТОРИЯ ВЫЧИСЛЕНИЙ:")
        print("="*50)
        if os.path.exists(self.history_file):
            with open(self.history_file, 'r', encoding='utf-8') as f:
                print(f.read())
        else:
            print("История пуста")
        print("="*50)
    
    def calculate(self, expr):
        """Вычисляет выражение"""
        try:
            # Безопасное вычисление
            result = eval(expr.replace('×', '*').replace('÷', '/'))
            
            # Сохраняем
            self.save_history(expr, result)
            
            return result, f"✅ Сохранено в файл"
        except ZeroDivisionError:
            return None, "❌ Ошибка: деление на ноль"
        except Exception as e:
            return None, f"❌ Ошибка: {str(e)}"
    
    def run(self):
        """Основной цикл программы"""
        print("="*50)
        print("КОНСОЛЬНЫЙ КАЛЬКУЛЯТОР")
        print("="*50)
        print("Введите выражение (например: 25 + 15)")
        print("Или команды: history, clear, exit")
        print("="*50)
        
        while True:
            try:
                user_input = input("\n>>> ").strip()
                
                if user_input.lower() == 'exit':
                    print("Выход из программы")
                    break
                elif user_input.lower() == 'history':
                    self.show_history()
                elif user_input.lower() == 'clear':
                    with open(self.history_file, 'w', encoding='utf-8') as f:
                        f.write("История вычислений:\n")
                    print("✅ История очищена")
                elif user_input:
                    result, message = self.calculate(user_input)
                    if result is not None:
                        print(f"Результат: {result}")
                        print(message)
                    else:
                        print(message)
                else:
                    print("Введите выражение или команду")
                    
            except KeyboardInterrupt:
                print("\nВыход из программы")
                break
            except Exception as e:
                print(f"Ошибка: {str(e)}")

# Запуск
calc = ConsoleCalculator()
```
